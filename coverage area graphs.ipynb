```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import math

l = []
sensor_positions = []
sensor_radii = []
bol = True
sensor = 1
a = int(input("Enter area in which sensors to be placed in sq meters: "))
while bol:
    r = float(input("Enter radius of sensor {}: ".format(sensor)))
    x, y = map(float, input("Enter x and y coordinates (comma-separated): ").split(','))
    coverage = 3.14 * r * r
    l.append(coverage)
    sensor_positions.append((x, y))
    sensor_radii.append(r)
    i = input("Do you want to continue? Enter yes or no: ")
    if(i == "no"):
        bol = False
    else:
        sensor += 1

s = 0
for i in range(len(l)):
    s += l[i]

print("Covered Area: {}".format(s))
print("Not Covered Area: ", a - s)
print("Total Number of Sensors Placed: {}".format(sensor))

# Plotting the data
fig, ax = plt.subplots()

for i in range(len(l)):
    # Draw sensor coverage area
    circle = patches.Circle(sensor_positions[i], sensor_radii[i], fill=False)
    ax.add_patch(circle)
    # Draw sensor point
    plt.plot(sensor_positions[i][0], sensor_positions[i][1], 'ro')
    
x_limit = math.sqrt(a)
y_limit = math.sqrt(a)
plt.xlim(0, x_limit)
plt.ylim(0, y_limit)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sensor Coverage')
plt.show()
```
